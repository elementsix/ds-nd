# 分类变量合并为一列去除重复项的问题解决

````python
#eg
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(10, 4),columns=list('abcd'))
df['piple'] = ['piple','none','piple','none','ddddd','none','piple','none','none','none']
df['floor'] = ['floor','none', 'floor','none','none','floor','none','none','floor','none']
df['ddddd'] = ['none','ddddd','none','none','ddddd','none','none','ddddd','none','none']
df
````

## solution1:采用melt函数  wide data 转换为long data，去除重复

````python
df1 = df.copy()
df1 = pd.melt(df1, id_vars=list('abcd'), var_name='dogs', value_name='stage')
df1
````

去除重复项

````python
df1_val = df1[df1.stage != 'none']
df1_val
````

````python
df1_none = df1[df1.stage == 'none']
df1_none
````

````python
df1_none_dup = df1_none.drop_duplicates(subset='a')
df1_none_dup
````
